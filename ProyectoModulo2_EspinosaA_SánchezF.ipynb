## Objetivo
Se requiere determinar si el sistema oscilante dado resulta ser  crticamente amortiguado, sobreamortiguado y subamortiguado
### Objetivos Especficos
Se requiere saber el tiempo en el que la masa pas por la posicin de equilibrio en el siguiente sistema y se necesita determinar el tiempo por el que la masa alcanza su desplazamiento extremo desde la posicin de equilibrio y su posicin en ese tiempo

## Modelo

Este modelo se le adjunta a un resorte con su extremo superior unido a un soporte fijo al cual se le acopla un contrapeso.

m= masa del contrapeso

Pero la ecuacin anterior se corrige aadiendo un trmino para la friccin el cual ser representado por FF

m= masa del contrapeso

K= Constante del resorte

## Problema
Una masa que pesa 4 lb se une a un resorte con una constante de muelle de 2 lb/ft. El medio ofrece una fuerza de amortiguamiento numricamente igual a la velocidad del instantneo, la masa se libera desde un punto situado un pie arriba de la posicin de equilibrio con velocidad descendente de 8 ft/s. Desarrolle la ecuacin para encontrar su grafica de velocidad y posicin.

## Simulacin 
Los datos proporcionados por el problema anterior son

W= 4lb (Peso)
K= 2lb/ft
B= 1lb*s/ft
Xo= -1 Se define como la posicin inicial de nuestro sistema (Al ser negativo nos dice que se encuentra arriba de la posicin de equilibrio
Xo= 8ft
Los datos anteriores deben de estar representados lo ms cercanamente posible a la ecuacin del modelo la cual es mx+Bx+kx=0
El primer paso es convertir el peso dado a masa:
M= W/g= 4lb/32ft (gravedad representada en el sistema ingles) = 1/8 slug (unidad de masa en el sistema lb/gravedad)

```python
# se importan las librerias necesarias a utilizar
import matplotlib.pyplot as plt        
import numpy as np                      
from scipy.integrate import odeint   
import matplotlib as mpl
%matplotlib inline

#%% se definen los datos dados por el problema

m = 1/8        # la masa es 1/8 ya que se convierte de lb a slugs         
k = 2                                                  
B = 1                                        

y0 = [-1, 8]     # condiciones iniciales                       
                     
# Funcin f(t,y) 
def armonico(y, t, k, m, B):
    y1 = y[0]
    y2 = y[1]
    return [y2, -k * y1 / m - B / m * y2]
         

# se definen los parametros necesarios para la grfica
tt = np.linspace(0, 3, 100)   

yy = odeint(armonico, y0, tt, args=(k, m, B))

yy.shape

y1 = yy[:,0]
y2 = yy[:,1]

label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 

plt.figure(figsize=(8,6))
plt.plot(tt, y1, 'b', label='posicin')
plt.plot(tt, y2, 'r', label='velocidad')
plt.legend(loc='best')
plt.xlabel('tiempo')
plt.ylabel('X(t)')
plt.grid()
plt.show()

```

## Conclusin:
Logramos definir el sistema anterior como un sistema sobreamortiguado por lo cual se cumplieron  los objetivos, obtuvimos este resultado por medio de la grfica las cual efectivamente nos muestra la posicin extrema de equilibro en el tiempo determinado y nos dice que el problema al tener races reales no repetidas el coeficiente de friccin es muy grande por lo cual el sistema se detendr de una manera ms rpida de lo que realmente debera.
